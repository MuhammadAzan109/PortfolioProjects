```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model
```

```python
dataset = pd.read_csv(r'H:\UET\Course Folder\BS\Course Material\Machine Learning\Lab Work\Datasets\Multiple Regression (Lab 3).csv')
```

```python
dataset
```

```python
sns.lmplot(x='Adv', y='Sale', data=dataset)
sns.lmplot(x='QC', y='Sale', data=dataset)
```

```python
Reg_Model = linear_model.LinearRegression()
```

```python
Reg_Model.fit(dataset[['Adv','QC']], dataset['Sale'])
```

```python
intercept = Reg_Model.intercept_
coefficients = Reg_Model.coef_
```

```python
print("Intercept (b):", intercept)
print("Coefficients(w1, w2):", coefficients)
```

```python
new_advertising = np.array([40, 42, 49])
new_quality_control = np.array([8, 10, 15])
```

```python
x_new = np.column_stack((new_advertising, new_quality_control))
```

```python
x_new
```

```python
predicted_sales = Reg_Model.predict(x_new)
```

```python
print("Predicted sales:", predicted_sales)
```

```python
from sklearn.metrics import r2_score
```

```python
pred = Reg_Model.predict(dataset[['Adv','QC']])
pred = np.vstack(pred)               # print the value in one column
pred
```
